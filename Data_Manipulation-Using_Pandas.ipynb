The data for this practice is downloaded from <a href="https://www.kaggle.com/CooperUnion/cardataset/download">CarData</a>

```python
data = pd.read_csv('Cardata.csv') 
data.head(n=3)
```

```python
data_sorted = data.sort_values(by = 'Engine HP')
data_sorted.head(n=3)
```

```python
data_groupby = data.groupby(data.Year).agg({'Popularity':['first','min','max','last']})
data_groupby.head(n=3)
```

```python
data_transform = data_groupby.transform(func=['sqrt'])
data_transform.head()
```

```python
data_transform = data_groupby.transform(lambda x:x+100)
data_transform.head()
```
