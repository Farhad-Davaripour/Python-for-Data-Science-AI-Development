```python
import yfinance as yf
import pandas as pd
```

```python
apple = yf.Ticker('AAPL')
google = yf.Ticker('GooG')
```

```python
google_info = apple.info
google_info = google.info
```

```python
apple_price_data = apple.history(period='max')
google_price_data = google.history(period='max')
```

```python
apple_price_data.reset_index(inplace=True)
google_price_data.reset_index(inplace=True)
```

```python
apple_price_data.plot(x='Date', y='Close', title='Apple History Stock Price Data', ylabel = 'stock price $', figsize=(8,5))
google_price_data.plot(x='Date', y='Close', title='Apple History Stock Price Data', ylabel = 'stock price $', figsize=(8,5))
```

```python
apple.earnings.plot(xlabel = 'Date', ylabel='divident $',figsize=(8,5))
google.earnings.plot(xlabel = 'Date', ylabel='divident $',figsize=(8,5))
```
