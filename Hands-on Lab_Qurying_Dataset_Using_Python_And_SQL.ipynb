```python
# The SCHOOLS table contains a large number of columns. How many columns does this table have?
%sql select count(*) from syscat.columns where tabname=upper('ChicagoPublicSchool')
```

```python
# Now retrieve the the list of columns in ChicagoPublicSchool table and their column type (datatype) and length.
%sql select colname, typename, length from syscat.columns where tabname = 'CHICAGOPUBLICSCHOOL'
```

```python
# How many Elementary Schools are in the dataset?¶
%sql select count(*) as count_school_ES from CHICAGOPUBLICSCHOOL where school='ES';
```

```python
# What is the highest Safety Score?
%sql select max(SAFETY_SCORE) as max_safety_score from CHICAGOPUBLICSCHOOL
```

```python
# What are the top 10 schools with the highest "Average Student Attendance"?
%sql select school, Average_Student_Attendance from ChicagoPublicSchool order by Average_Student_Attendance desc nulls last limit 10
```

```python
# Retrieve the list of 5 Schools with the lowest Average Student Attendance sorted in ascending order based on attendance
%sql select school, Average_Student_Attendance \
from ChicagoPublicSchool order by Average_Student_Attendance nulls last fetch first 5 rows only
```

```python
# Now remove the '%' sign from the above result set for Average Student Attendance column
%sql select replace(Average_Student_Attendance,'%','') as First_5_Average_Student_Attendance from ChicagoPublicSchool\
order by Average_Student_Attendance\
fetch first 5 rows only
```

```python
# Which Schools have Average Student Attendance lower than 70%?
%sql select school, Average_Student_Attendance from ChicagoPublicSchool\
where Decimal(replace(Average_Student_Attendance,'%',''))<70\
order by Average_Student_Attendance Desc;
```

```python
# Get the total College Enrollment for each Community Area
%sql select COMMUNITY_AREA_NAME , sum(College_Enrollment) as sum_of_college_enrollment\
from ChicagoPublicSchool group by COMMUNITY_AREA_NAME;
```

Get the 5 Community Areas with the least total College Enrollment sorted in ascending order

```python
%%sql
select COMMUNITY_AREA_NAME , sum(College_Enrollment) as sum_of_college_enrollment
from ChicagoPublicSchool
group by COMMUNITY_AREA_NAME
order by sum_of_college_enrollment
fetch first 5 rows only;
```

List 5 schools with lowest safety score.

```python
%%sql
select NAME_OF_SCHOOL , sum(Safety_Score) as Sum_Safety_Score
from ChicagoPublicSchool 
group by NAME_OF_SCHOOL
order by Sum_Safety_Score
Fetch first 5 rows only;
```

Get the hardship index for the community area which has College Enrollment of 4368

```python
%%sql
select community_area_name, hardship_index from SOCIOECONOMIC_INDICATOR
where upper(community_area_name) = upper((select community_area_name from ChicagoPublicSchool where College_Enrollment=4368));
```

```python
%%sql
select indicator.community_area_name , hardship_index 
from SOCIOECONOMIC_INDICATOR indicator, ChicagoPublicSchool school
where upper(indicator.community_area_name) = upper(school.community_area_name)
and College_Enrollment=4368
```

Get the hardship index for the community area which has the school with the highest enrollment

```python
%%sql
select indicator.community_area_name , hardship_index 
from SOCIOECONOMIC_INDICATOR indicator, ChicagoPublicSchool school
where upper(indicator.community_area_name) = upper(school.community_area_name)
and College_Enrollment=(select max(College_Enrollment) from ChicagoPublicSchool)
```
