### An example of web scraping
This example is a part of 'Python for Data Science, AI & Development' course offered by coursera.org   
For the solution, the code skeleton provided in the lab is employed.

```python
url = "https://en.wikipedia.org/wiki/Population_of_Canada"
```

```python
import requests
from bs4 import BeautifulSoup as bs
r = requests.get(url).text
soup = bs(r, 'html.parser')
```

```python
data = soup.find_all('table')
for i, content in enumerate(data):
    if 'French population only' in str(content):
        index_num = i
```

```python
tables = data[index_num].tbody.find_all('tr')
```

```python
for title in tables:
    titles = title.find_all('th')
    if titles !=[]:
        for i in range(len(titles)):
            print(titles[i].contents[0].strip())
```

```python
import pandas as pd
Year = []
Area_colony = []
population = []
Notes = []
for element in tables:
    elements = element.find_all('td')
    if elements !=[]:
        Year.append(elements[0].text)
        Area_colony.append(elements[1].text.strip())
        population.append(elements[2].text.strip())
        Notes.append(elements[3].text.strip())

df = pd.DataFrame({'Year':Year,"Area_colony":Area_colony,'population':population,'Notes':Notes})
df.head(n=10)
```

__Here is an easy way to tabulate the data using Pandas__

```python
table_index_num = pd.read_html(str(data[index_num]), flavor='bs4')[0]
table_index_num
```

__Also we can tabulate all the tables directly from the url__

```python
table_list = pd.read_html(url, flavor = 'bs4')
```

__Finally we can even filter the table that we are interested in using pandas.read_html__

```python
df_readhtml_match = pd.read_html(url, match='French population only', flavor='bs4')[1]
df_readhtml_match
```
